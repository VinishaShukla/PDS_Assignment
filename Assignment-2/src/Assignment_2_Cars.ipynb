```python
import pandas as pd

# Read the dataset
df_automobiles = pd.read_csv('/Users/vinishashukla/Documents/PDS/PDS_Assignment/PDS_Assignment/Assignment-2/raw_data/train.csv')
df_automobiles
```

```python
# Checking for missing values in the entire dataset
missing_vals = df_automobiles.isnull().sum()

# Display the number of missing values for columns with numeric datatype
print(missing_vals)
```

```python
# Identify and count missing values in columns that have character datatype
missing_vals_char = df_automobiles.select_dtypes(include=['object']).isnull().sum()

# Display the number of missing values in columns that have character datatype
print(missing_vals_char)
```

```python
# Remove the units from some of the attributes and only keep the numerical values
df_automobiles['Mileage'] = df_automobiles['Mileage'].str.replace('[^0-9.]', '', regex=True).astype(float)
df_automobiles['Engine'] = df_automobiles['Engine'].str.replace('[^0-9.]', '', regex=True).astype(float)
df_automobiles['Power'] = df_automobiles['Power'].str.replace('[^0-9.]', '', regex=True).astype(float)
df_automobiles['New_Price'] = df_automobiles['New_Price'].str.replace('[^0-9.]', '', regex=True).astype(float)

# Print the summary of the data
print(df_automobiles.describe())
```

```python
# Drop the first column which is just a row number
df_automobiles = df_automobiles.iloc[:, 1:]

# Drop the "New_Price" column because it has more than 50% missing values
df_automobiles = df_automobiles.drop('New_Price', axis=1)

# Impute missing values with the mean for numeric columns
df_automobiles.fillna(df_automobiles.mean(), inplace=True)

# Round all numeric values to two decimal places
numeric_columns = df_automobiles.select_dtypes(include=['float64', 'int64']).columns
df_automobiles[numeric_columns] = df_automobiles[numeric_columns].round(2)

# Impute missing values with the mode for character columns
df_automobiles.fillna(df_automobiles.mode().iloc[0], inplace=True)

# Verify that missing values have been correctly imputed
missing_vals_handled = df_automobiles.isnull().sum()
print(missing_vals_handled)
```

```python
# Write the cleaned data to a new file
df_automobiles.to_csv('/Users/vinishashukla/Documents/PDS/PDS_Assignment/PDS_Assignment/Assignment-2/clean_data/clean_cars_data.csv', index=False)
```

```python
import pandas as pd

# Load the dataset
df_cars = pd.read_csv('/Users/vinishashukla/Documents/PDS/PDS_Assignment/PDS_Assignment/Assignment-2/results/cars_current_age.csv')

# Select specific columns
selected_columns = df_cars[['Name', 'Location', 'Year', 'Kilometers_Driven', 'Fuel_Type', 'Transmission', 'Owner_Type', 'Mileage', 'Engine', 'Power', 'Seats', 'Price']]

# Filter rows based on condition
filtered_data = df_cars[df_cars['Fuel_Type'] == 'Petrol']

# Rename columns
renamed_columns = df_cars.rename(columns={'Transmission': 'TransType', 'Owner_Type': 'OwnerType'})

# Mutate (create) a new column based on existing columns
df_cars['Power_per_Cylinder'] = df_cars['Power'] / df_cars['Seats']

# Arrange rows based on a column
sorted_data = df_cars.sort_values(by='Year', ascending=False)

# Summarize data using group by operation
grouped_data = df_cars.groupby('Location').agg({'Price': 'mean', 'Kilometers_Driven': 'sum'})

# Display the results
print("Selected Columns:")
print(selected_columns.head())


```
